```python
contagemDeLinguas = tmdb['original_language'].value_counts().to_frame().reset_index()
contagemDeLinguas.columns = ['original_language', 'total']
contagemDeLinguas.head()
```

```python
sns.barplot(x= 'contagemDeLinguas', y = 'total', data = contagemDeLinguas)
```

```python
sns.catplot(x = 'original_language', kind='count', data = tmdb)
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua['en']
total_resto = total_geral - total_de_ingles
```

```python
dados = {
    'lingua' : ["ingles", "outros"],
    'total' : [total_de_ingles, total_resto]}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(x ='lingua', y = 'total', data =dados)
```

#Remoção da lingua inglesa devido a descrepância de sua presença quando comparada as outras linguas, dificultando assim o estudo como um todo.

```python
total_por_lingua_de_outros_filmes = tmdb.query("original_language != 'en'").original_language.value_counts()
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
```

```python
sns.catplot(x = 'original_language', kind='count',
            data = filmes_sem_lingua_original_em_ingles,
            aspect=2,
            palette= 'GnBu_d',
            order= total_por_lingua_de_outros_filmes.index
            )
```

```python
for lingua, valor in total_por_lingua_de_outros_filmes.items():
    print(f"Língua: {lingua}, Valor: {valor}")
```
